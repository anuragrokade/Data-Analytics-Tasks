```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dt= pd.read_csv('Dataset.csv')
```

```python
restaurants_count = dt.groupby('City')['Restaurant ID'].count()

city_with_max_restaurants = restaurants_count.idxmax()

print('City with highest no. of restaurants : ', city_with_max_restaurants)
```

```python
avg_rating = dt.groupby('City')['Aggregate rating'].mean()
```

```python
avg_rating.plot(kind='bar', figsize=(30,10))
plt.xlabel('City',fontweight='bold')
plt.ylabel('Average Rating',fontweight='bold')
plt.title('Average rating for Restaurants in each city',fontweight='bold')
plt.show()
```

```python
city_with_max_avg_rating = avg_rating.idxmax()
print('City with highest average rating :',city_with_max_avg_rating )
```
