```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset.csv")
```

```python
df
```

```python
price_range_count = df["Price range"].value_counts().sort_index()
```

```python
total_restaurants = len(df)
```

```python
percentage_per_price_range = (price_range_count / total_restaurants)*100
```

```python
plt.bar(price_range_count.index,price_range_count.values)
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.title("Distribution of Price Range")
plt.show()
```

```python
print("Percentage of restaurants in each price range category :")
print(percentage_per_price_range)
```
